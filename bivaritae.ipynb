Step 0: Setup and Data Loading

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Loading  dataset
try:
    df = pd.read_csv('Bivariate analysis\loan_sanction_train.csv')
    print("Dataset loaded successfully!")
    print("Shape of the dataset:", df.shape)
except FileNotFoundError:
    print("Error: 'loan_sanction_train.csv' not found. Please check the file path.")
```

The bivariate analysis will be structured around these three types of pairings:

Categorical vs. Categorical

Numerical vs. Numerical

Numerical vs. Categorical

Categorical Variables: Gender, Married, Dependents, Education, Self_Employed, Credit_History, Property_Area, Loan_Status.

Numerical Variables: ApplicantIncome, CoapplicantIncome, LoanAmount, Loan_Amount_Term.


```python
# 1. Using Crosstabulation
# This table shows the raw counts of loan approvals and rejections per property area
print("Crosstab of Property_Area vs. Loan_Status:")
crosstab_result = pd.crosstab(df['Property_Area'], df['Loan_Status'])
print(crosstab_result)

# 2. Using a Count Plot for Visualization
plt.figure(figsize=(8, 6))
sns.countplot(x='Property_Area', hue='Loan_Status', data=df, palette='viridis')
plt.title('Loan Status by Property Area', fontsize=15)
plt.xlabel('Property Area')
plt.ylabel('Count of Applicants')
plt.show()
```

```python
# 1. Using a Scatter Plot
plt.figure(figsize=(10, 7))
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=df, alpha=0.5)
plt.title('Applicant Income vs. Loan Amount', fontsize=15)
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

```python
# 2. Using a Correlation Heatmap for all numerical variables
# Select only the numerical columns for the correlation matrix
numerical_cols = df.select_dtypes(include=['number'])
correlation_matrix = numerical_cols.corr()

plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables', fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Loan_Status', y='ApplicantIncome', data=df, palette='pastel')
plt.title('Applicant Income Distribution by Loan Status', fontsize=15)
plt.xlabel('Loan Status')
plt.ylabel('Applicant Income')
plt.show()
```
